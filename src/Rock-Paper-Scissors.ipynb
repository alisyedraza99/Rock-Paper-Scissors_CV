```python
from tensorflow.python.keras.optimizers import Adam
#from tensorflow.python.keras.models import Model,load_model
#from tensorflow.python.keras.applications.inception_v3 import InceptionV3
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense,Input,Flatten, Dropout, Conv2D, MaxPooling2D
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from tensorflow.python.keras.callbacks import EarlyStopping,ReduceLROnPlateau
```

```python
num_classes = 4
```

```python
p_model = Sequential()
p_model.add(Conv2D(64, (3,3), activation='relu', input_shape=(150, 150, 3)))
p_model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
p_model.add(MaxPooling2D(2, 2))
p_model.add(Conv2D(128, (3,3), activation='relu', padding='valid'))
p_model.add(Conv2D(128, (3,3), activation='relu', padding='same'))
p_model.add(MaxPooling2D(2, 2))
#p_model.add(InceptionV3(include_top=False, weights='imagenet', input_shape=(300, 300, 3), pooling=None, classes=num_classes))
p_model.add(Flatten()) #flatten out the model
p_model.add(Dropout(0.6))
#p_model.add(Dense(1024, activation='relu'))
#p_model.add(Dropout(0.5))
p_model.add(Dense(512, activation='relu'))
p_model.add(Dense(num_classes, activation='softmax'))

#p_model.layers[0].trainable = False
```

```python
p_model.compile(optimizer='adam', 
                     loss='categorical_crossentropy', 
                     metrics=['accuracy'])
```

```python
#Get number of images

path = '../input/rps-updated/Rock-Paper-Scissors/'
folders = os.listdir(path)
folders.sort()
train_imgs = 0
val_imgs = 0
folders = [path + f + "/" for f in folders] 
print(folders)
train_set = os.listdir(folders[1])
train_set = [folders[1] + f + "/" for f in train_set]
print(train_set)

#names for test and val set are not used correctly in this dataset so re-adjusting
val_set = os.listdir(folders[0])
val_set = [folders[0] + f + "/" for f in val_set]
print(val_set)

for hand in train_set:
    train_imgs += len(os.listdir(hand))
for hand in val_set:
    val_imgs += len(os.listdir(hand))
print(train_imgs, val_imgs)
```

```python
p_model.summary()
BATCH_SIZE = 25
STEPS_PER_EPOCH = train_imgs//BATCH_SIZE
VAL_STEPS = val_imgs//BATCH_SIZE
image_size = (150, 150)

train_data_gen = ImageDataGenerator(rescale=1./255,
                                    horizontal_flip=True,
                                    vertical_flip=False,
                                    brightness_range=(0.5,1.5),
                                    rotation_range=10) 
val_data_gen = ImageDataGenerator(rescale=1./255) #we do not augment validation data

train_generator = train_data_gen.flow_from_directory(
        folders[1],
        target_size=image_size,
        batch_size=BATCH_SIZE)

val_generator = val_data_gen.flow_from_directory(
        folders[0],
        target_size=image_size,
        batch_size=BATCH_SIZE)

p_model.fit_generator(train_generator,
                    validation_data=val_generator,
                    steps_per_epoch=STEPS_PER_EPOCH,
                    validation_steps=VAL_STEPS,
                    epochs=20)
```

```python
import cv2
import matplotlib.pyplot as pli

label_map = (train_generator.class_indices)
labels = {v: k for k, v in label_map.items()} 
print(label_map)
print(p_model)
p_model.save("rps.h5")

folders = os.listdir('../input/rock-paper-scissors-dataset/rock-paper-scissors/Rock-Paper-Scissors/validation/')
plt.figure(figsize=(30, 20))
i = 0
for f in folders:
    path = '../input/rock-paper-scissors-dataset/rock-paper-scissors/Rock-Paper-Scissors/validation/' + f
    img = cv2.resize(cv2.imread(path, cv2.IMREAD_COLOR), (150, 150), interpolation=cv2.INTER_CUBIC)
    new = np.expand_dims(img, axis=0)
    new = new[:,:,:, :3]
    result = p_model.predict(new)
    pred_index = result.argmax(axis=1)
    print(pred_index)
    plt.subplot(11+1, 3, i+1)
    plt.title('This is ' + labels[pred_index[0]])
    imgplot = plt.imshow(img)
    i += 1
plt.show()
#filename = "rps_model.pkl"  

#with open(filename, 'wb') as file:  
#    pickle.dump(p_model, file)
```
